```python
import pandas as pd
from scipy.stats import hypergeom as hg
```

```python
# nerve 1 data from comparison 1 paper can be passed to the following function to obtain the p value as reported in the paper table
# hg.sf(x-1, M, n, N)
# M is the population size (previously N)
# n is the number of successes in the population (previously K)
# N is the sample size (previously n)
# X is still the number of drawn “successes”
```

```python
# No razor, no mouse, no virulent
x = 11
M = 52752
n = 44
N = 796
expected = (N * n ) / M
fe = x / expected
p = hg.sf(x-1, M, n, N)
print(f'Expected: {expected}')
print(f'Fold-enrichment: {fe}')
print(f'p-value: {p}')
```
